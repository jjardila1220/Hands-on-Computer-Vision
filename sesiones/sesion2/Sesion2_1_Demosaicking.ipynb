## üìå Introducci√≥n

**Demosaicking** (o de-mosaicing) es un algoritmo utilizado en el procesamiento de im√°genes digitales para reconstruir im√°genes a todo color a partir de datos incompletos obtenidos por un sensor de imagen con una matriz de filtros de color (CFA), como el filtro Bayer.

üì∑ **C√°maras digitales modernas** usan sensores con CFA, lo que hace que el demosaicking sea crucial para obtener im√°genes visibles y de alta calidad. Las im√°genes RAW permiten que este proceso se haga v√≠a software en vez del firmware de la c√°mara.


## üì∏ **Formatos de im√°genes RAW m√°s comunes**

## üåà **2. Hagamos demosaicking** <a name="tema3"></a> üé®

**Objetivo:** Separar la imagen en sus canales RGB.

La funci√≥n `get_cfa_channel` debe devolver 0 en rojo, 1 en verde, y 2 en blue.
La funci√≥n `get_cfa_mask` debe devolver una matrix de tama√±o de la imagen original, donde cada valor corresponde a 0, 1 o 2

```python
# @title interpolaci√≥n bilineal
import matplotlib.pyplot as plt  # Importa Matplotlib para visualizaci√≥n
import numpy as np  # Importa NumPy para manejo de matrices

# Crear matrices 4x4 llenas de ceros (representando una imagen en escala de grises inicialmente)
green_matrix = np.zeros((4, 4))
blue_matrix = np.zeros((4, 4))
red_matrix = np.zeros((4, 4))

# Definir las posiciones donde hay p√≠xeles activos en cada canal de color
green_positions = [(0, 1), (1, 0), (1, 2), (2, 1)]
blue_positions = [(0, 0), (0, 2), (2, 0), (2, 2)]
red_positions = [(1, 1), (1, 3), (3, 1), (3, 3)]

# Asignar 1 en las posiciones correspondientes de cada matriz de color
for i, j in green_positions:
    green_matrix[i, j] = 1  # Activa los p√≠xeles verdes

for i, j in blue_positions:
    blue_matrix[i, j] = 1  # Activa los p√≠xeles azules

for i, j in red_positions:
    red_matrix[i, j] = 1  # Activa los p√≠xeles rojos

# Definir etiquetas para cada canal de color con nombres espec√≠ficos
green_labels = {(0, 1): 'G1', (1, 0): 'G4', (1, 2): 'G2', (2, 1): 'G3', (1, 1): 'Gx'}
blue_labels = {(0, 0): 'B1', (0, 2): 'B2', (2, 0): 'B3', (2, 2): 'B4', (1, 1): 'Bx'}
red_labels = {(1, 1): 'R1', (1, 3): 'R2', (3, 1): 'R3', (3, 3): 'R4', (2, 2): 'Rx'}

# Funci√≥n para graficar una matriz con su correspondiente mapa de calor y etiquetas
def plot_matrix(ax, matrix, cmap, labels):
    ax.imshow(matrix, cmap=cmap, vmin=0, vmax=1)  # Visualiza la matriz en escala de colores

    # Dibujar la cuadr√≠cula sobre la imagen
    for i in range(5):
        ax.plot([-0.5, 3.5], [i - 0.5, i - 0.5], color='black', linewidth=1.5)  # L√≠neas horizontales
        ax.plot([i - 0.5, i - 0.5], [-0.5, 3.5], color='black', linewidth=1.5)  # L√≠neas verticales

    # Agregar etiquetas en las celdas de la cuadr√≠cula
    for (i, j), text in labels.items():
        ax.text(j, i, text, ha='center', va='center', fontsize=14, color='gray')

    # Ocultar los ejes para una visualizaci√≥n m√°s clara
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_frame_on(False)

# Crear la figura con 3 subgr√°ficos (uno por cada canal de color)
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

# Graficar cada matriz con su respectivo color y etiquetas
plot_matrix(axes[0], green_matrix, 'Greens', green_labels)
plot_matrix(axes[1], blue_matrix, 'Blues', blue_labels)
plot_matrix(axes[2], red_matrix, 'Reds', red_labels)

# Ajustar el dise√±o para evitar superposici√≥n
plt.tight_layout()
plt.show()
```

```python
RED, GREEN, BLUE = 0, 1, 2


def get_cfa_channel(i, j):
    if i % 2 == 0 and j%2 == 0:
        return RED
    elif i % 2 == 0 and j % 2 == 1:
        return GREEN
    elif i % 1 == 1 and j % 2 == 0:
        return BLUE
    else:
        return RED

def get_cfa_mask(shape):
    a = np.zeros(shape)
    for i in range(shape[0]):
      for j in range(shape[1]):
        a[i, j] = get_cfa_channel(i, j)
    return a

cfa_mask = get_cfa_mask(image.shape)

r_mask, g_mask, b_mask = (cfa_mask == RED), (cfa_mask == GREEN), (cfa_mask == BLUE)

r, g, b = image * r_mask, image * g_mask, image * b_mask

plt.imshow(r + g + b)
plt.show()
```

‚úÖ **Siguiente paso:** Completa los huecos en la imagen interpolando los valores faltantes.
Esta vez, vamos a utilizar convoluciones y sus kernel de convoluci√≥n.

```python
# üìå Definir los filtros de interpolaci√≥n para cada canal.
G_KERNEL = np.array([[0, 1, 0],  # Interpolaci√≥n cruzada para el verde
                     [1, 0, 1],
                     [0, 1, 0]])

RB_KERNEL = np.array([[1, 0, 1],  # Interpolaci√≥n diagonal para rojo y azul
                      [0, 0, 0],
                      [1, 0, 1]])

# üìå Aplicar convoluci√≥n para interpolar los valores faltantes en cada canal.
r = convolve(r, RB_KERNEL, mode='nearest')  # Interpola el canal rojo
b = convolve(b, RB_KERNEL, mode='nearest')  # Interpola el canal azul
g = convolve(g, G_KERNEL, mode='nearest')   # Interpola el canal verde

# üìå Almacenar canales y t√≠tulos para graficarlos.
channels = [r, g, b]
titles = ['Red Channel', 'Green Channel', 'Blue Channel']

# üìå Graficar cada canal y su versi√≥n con zoom.
for i, channel in enumerate(channels):
    plt.subplot(3, 2, 2*i + 1)  # Imagen completa
    plt.imshow(channel, cmap='gray')
    plt.title(titles[i])
    plt.colorbar()

    plt.subplot(3, 2, 2*i + 2)  # Imagen con zoom en la regi√≥n [310:410, 1405:1505]
    plt.imshow(channel[310:410, 1405:1505], cmap='gray')
    plt.title(f'{titles[i]} (Zoomed)')
    plt.colorbar()

# üìå Ajustar distribuci√≥n de gr√°ficos y mostrar la figura.
plt.tight_layout()
plt.show()
```

## üéØ **3. Correcciones de color** <a name="tema4"></a> üåà

```python
import numpy as np
import matplotlib.pyplot as plt

# üìå Definimos el valor de gamma para la correcci√≥n gamma
gamma = 2.2  # üõ†Ô∏è Puedes cambiar este valor para ver c√≥mo afecta a la imagen

# üìå Apilamos los canales rojo, verde y azul para formar una imagen RGB
rgb = np.stack([r, g, b], axis=-1)  # Apila las matrices r, g, b en un solo array tridimensional

# üìå Evitar divisiones por cero en la normalizaci√≥n
min_val = np.min(rgb, axis=(0, 1), keepdims=True)  # M√≠nimo por canal, manteniendo dimensiones
max_val = np.max(rgb, axis=(0, 1), keepdims=True)  # M√°ximo por canal, manteniendo dimensiones

# üìå Asegurar que no haya divisiones por cero
range_val = max_val - min_val
range_val[range_val == 0] = 1  # Evita divisiones por cero

# üìå Normalizaci√≥n Min-Max por canal
rgb = (rgb - min_val) / range_val

# üìå Aplicamos la correcci√≥n gamma a la imagen RGB
rgb = np.power(rgb, 1.0 / gamma)

# üìå Asegurar que los valores est√©n en el rango correcto para `imshow`
rgb = np.clip(rgb, 0, 1)  # Recorta valores fuera de [0,1] para evitar errores en la visualizaci√≥n

# üìå Mostramos la imagen resultante
plt.imshow(rgb)
plt.axis('off')  # Ocultar ejes para una mejor visualizaci√≥n
plt.show()
```
